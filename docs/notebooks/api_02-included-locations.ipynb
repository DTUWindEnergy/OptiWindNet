## Included locations

```python
from optiwindnet.api import WindFarmNetwork
```

Data of several locations is included in optiwindnet repository (data folder). We can see the list as:

```python
from pathlib import Path

data_dir = Path('../../optiwindnet/data')
files = list(data_dir.glob('*'))  # Use '*.yaml' to filter by extension

for f in files:
    print(f.name)
```

Check if the file data is available and then instantial WindFarmNetwork (using .from_yaml or .from_pbf depending on the file format)

```python
Path('../../optiwindnet/data/Seagreen.osm.pbf').exists()
```

```python
file_path = '../../optiwindnet/data/Seagreen.osm.pbf'
wfn = WindFarmNetwork.from_pbf(filepath=file_path, cables=7)
```

```python
# set Matplotlib option to generate inline SVG
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
plt.rcParams['svg.fonttype'] = 'none'

# plot location
wfn.plot_location()
```
